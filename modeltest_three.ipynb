```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.read_csv('data/usethis.csv')
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
original = pd.read_csv('data/ZILLOWforMom.csv')
original = original.drop(columns=['Unnamed: 0', 'county', 'neighborhood_name', 'RealEstateCompany','is_new_construction', 'property_id','postal_code', 'price_reduced_date','listing_id', 'email', 'PrimaryNumber',
                                 'ListingAgent', 'unit', 'line', 'prop_status'])
original.head()
```

```python
#Reordering Dataframe based on interest of variables
original = original[['ListDate', 'LastUpdate', 'list_date', 'last_update', 'prop_type', 'price', 'baths', 'baths_full', 'baths_half', 'city', 'state_code',
                    'year_built', 'Property Size(Sq. Ft)', 'photo_count', 'rank']]
original.head()
```

```python
#original = original.drop(columns=['list_date', 'last_update'])
original.head()
```

```python
original['ListDate'] = pd.to_datetime(original.ListDate)
original['LastUpdate']=pd.to_datetime(original.LastUpdate)
```

```python
original.head()
```

```python
#Converting columns' objects into datetime objects for subrtraction. The difference is the most important factor for this study.
daysonMarket = original['LastUpdate']-original['ListDate']
daysonMarket.head()
```

```python
days_df = pd.DataFrame(daysonMarket)
days_df = days_df.rename(columns={0: 'Time'})
```

```python
original = original.merge(days_df, right_index = True, left_index = True)
original.head()
```

```python
original = original.drop(columns=['ListDate', 'LastUpdate'])
original.head()
```
